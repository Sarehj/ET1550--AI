```python
# Import libraries
import numpy as np
import pandas as pd
# import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset = pd.read_csv('ionosphere.csv', header=None)
```

```python
# Check the description of the dataset
dataset.describe()
```

```python
# Check the shape of the dataset
dataset.head()
```

```python
# Separate input from output
y = dataset.values[:, 34]
X = dataset.values[:, 0:34]
```

```python
# Split datasets into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
```

```python
n_features = X_train.shape[1]
```

```python
import tensorflow as tf
from tensorflow.keras import layers
# from tensorflow.keras import regularizers

model_NN = tf.keras.Sequential([
    layers.Dense(12, activation='relu', input_shape=(n_features,)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
# compile the model
model_NN.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# fit the model
model_NN.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
results = model_NN.evaluate(X_test, y_test)
print("test loss, test acc:", results)
```
